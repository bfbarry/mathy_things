```python

def sum_digits(number):
    str_num = str(number)
    root_array = [int(i) for i in str_num]
    digital_root = sum(root_array)
    return digital_root
    
def dr_degree_n(number, degree = 'full'):
    """
    returns full digital root unless degree is stated
    """
    root = sum_digits(number)
    if degree == 'full':
        while len(str(root)) != 1:
            root = sum_digits(root)
    else:
        while (degree != 1):
            if degree < 1:
                root = number
                break
            root = sum_digits(root)
            degree-=1
                
    return root
```

```python
def compare_polynomial_roots(length, exp_lim):
    all_roots = []
    x_axis = np.arange(1,length + 1)
    exp_range = np.arange(1, exp_lim + 1)
    
    for exp in exp_range:
        roots = [dr_degree_n(i**exp) for i in x_axis]
        all_roots.append(roots)
    
    #Plotting
    plt.figure(figsize=(20,10))
    for exp in exp_range:
        plt.plot(x_axis, all_roots[exp-1], 'o-', label = 'x^' + str(exp))
    plt.legend(loc='best')
    plt.show()
    
    return all_roots
```

```python
def plot_digital_roots(num_array, verbose = True):
    plt.figure(figsize=(20,10))
    x_axis = np.arange(len(num_array))
    roots = np.array([dr_degree_n(i) for i in num_array])
    plt.plot(x_axis, roots, 'o-')
    if verbose == True:
        print('unique vals:', np.unique(roots))
    return roots
```
