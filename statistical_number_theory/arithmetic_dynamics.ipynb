```python
import matplotlib.pyplot as plt
import numpy as np
import math
from IPython.display import clear_output
```

```python
plt.figure(figsize=(10,10))
f = lambda x: (x**3 + x+1)**0.5
def plot_frame(f, color='k',width=3,res=5000):
    xs = np.linspace(-5,4,res)
    plt.plot(xs, f(xs), color=color,linewidth=width)
    plt.plot(xs, -1*f(xs), color=color,linewidth=width)
    plt.ylim([-5,5])
    
def plot_sequence(a = [-2,2], b = [-0.5,2], noise=False):
    a = np.flip(np.linspace(a[0],a[1],100))
    b = np.flip(np.linspace(b[0],b[1],100))
    g = np.random.normal(scale=0.05,size=100)
    for a0, b0,n  in zip(a,b,g):
        if noise:
            f = lambda x: (x**3 + (n+a0)*x + b0+n)**0.5
        else:
            f = lambda x: (x**3 + a0*x+b0)**0.5
        plot_frame(f)
        clear_output(wait=True)
        plt.show()
        
plot_sequence(noise=1)
```
