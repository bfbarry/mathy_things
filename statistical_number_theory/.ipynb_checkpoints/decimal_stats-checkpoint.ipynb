# π

This'd be cool to vis in d3.js like a tub of balls gradually setting to a point

```python
for i, sze in enumerate (np.arange(100, 2100, 100)):
    if i == 0:
        plt.figure(figsize = [10,10])
    pi_i = sorted(generators.pi_dig(int(sze)))
    plt.subplot(5,4,i+1)
    plt.hist(pi_i, color = '#8142f5', alpha = .75)
plt.show()
```
