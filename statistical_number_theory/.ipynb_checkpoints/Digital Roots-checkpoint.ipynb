```python
def gen_n_primes(n=10000,fname='primes')
    primes = []
    i=0
    while len(primes) < n: 
        if i>1: 
            for j in range(2,i): 
                if(i % j==0): 
                    break
            else: 
                primes.append(i)
        i+=1
    if fname:
        with open(f'./data/{fname}.json', 'w') as f:
            json.dump(primes,f)
    return primes
```

```python
#numbers = np.random.randint(1000,size=(10000))]
roots = np.array([digital_roots.dr(n**2) for n in primes]).reshape(100,100)

plt.figure(figsize = (20,8))
plt.subplot(1,2,1)
plt.imshow(roots, cmap = plt.cm.get_cmap('YlGn')); plt.colorbar()
plt.subplot(1,2,2)
plt.hist(roots.reshape(10000)); plt.locator_params(axis='x', nbins=10)
```

```python
plt.figure()
for e in range(1,6):
    roots = np.array([digital_roots.dr(n**e) for n in primes]).reshape(100,100)
    pca_fit = PCA(100).fit(roots)
    evals = pca_fit.explained_variance_ratio_
    plt.loglog(np.arange(90),evals[:90])
```

```python
d_roots = []
for e in range(1,19): #over range of exponents
    roots_e = []
    for i in [n**e for n in range(10000)]:
        roots_e.append(digital_roots.dr(i))
    roots_e = np.array(roots_e)
    d_roots.append(np.reshape(roots_e,(100,100)))

colors = list(iter(plt.cm.rainbow(np.linspace(0, 1, 10))))
plt.figure(figsize=(16,20))    
for i in range(18):
    plt.subplot(6,3,i+1)
    plt.imshow(d_roots[i],cmap=plt.cm.rainbow)
    plt.axis('off')
    #plt.tight_layout()
    
plt.savefig('digital root patterns', dpi=200)
```

# Observations:

Here I am interpreting the digits of pi's decimal as the digits of a big integer.  In the above figure I calculated a running digital root for each of the first 1000 digits of pi.  So at each decimal place of pi a new integer is created, whose digital root is calculated.  So the last value is the actual digital root of this digit.  

- So it might be interesting to look at these time "series" of the digital roots of elements throughout a set.  We've looked at the digital roots of certain exponents, of primes, and now this weird representation of the digits of pi.  
- What differences in time series shape do we observe?

There seem to be certain patterns in how the time series manifest themselves
- We see repeating roots in the pi series, but not in primes and certain exponents.  Why is that?
- Other patterns include slants, and magnitude of divergence from the last root 
- granger causality?
