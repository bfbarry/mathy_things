````python



def mpl_set_cmap(colors, num):
    cmap = mpl.colors.LinearSegmentedColormap.from_list('mycmap', colors)(np.linspace(0, 1, num))
    mpl.rcParams['axes.prop_cycle'] = cycler.cycler('color', cmap)
        
def plot_continuous_power_curves(domain:[int, int]):
    
    
    things = []
    domain = np.linspace(*domain, 60)
    for base in domain:
        things.append([np.power(base, exponent) for exponent in range(100)])
    things = np.array(things)
    mpl_set_cmap(['orange', 'purple'], len(things))

    plt.loglog(things.T)
    
    
def plot_discrete_curves(domain):
    things = []
    for base in domain:
        things.append([np.power(base, exponent) for exponent in range(100)])
    things = np.array(things)
    mpl_set_cmap(['orange', 'purple'], len(things))

    plt.loglog(things.T)

plt_domains = [[-0.5, 0.5], [0,4], [-4,4], [-4,-2]]
# plt_domains =[[0,4]]
plt.figure(figsize=(10,10))
for i, dom in zip(range(4), plt_domains):
    plt.subplot(2,2, i+1)
    plot_continuous_power_curves(dom)
    plt.title(dom)
````
