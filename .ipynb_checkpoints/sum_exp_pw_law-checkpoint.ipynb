```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

When drawing from gaussians, only want vals > 0. Best way to do this?
- turn all vals positive? filter negative values (given mean, std ensuring mostly pos)?

How to determine the range classify result as a power law? e.g. for a,t = 20, 11, an exponential function looks quite power law on domain (0,2.5)
    - this is a function of a and t's orders of magnitude 

```python
def add_exps(num=2, distr = 'normal', dc=2, dom=(2.5,100)):
    ''' 
    num: number of exponentials
    distr: 
    dc: dc * num is size of random number set to draw from
    dom = (stop, num_points)
    '''
    exps = [] # exp functions
    params = [] # just the params
    if distr is 'normal':
        # algo to filter negative draws
        a_rands = []
        t_rands = []
        c=0
        while c < num*dc:
            try_a =np.random.normal(loc=20, scale=1.0)
            try_t = np.random.normal(loc=10, scale=1.0)
            if try_a > 0 and try_t > 0:
                a_rands.append(try_a) 
                t_rands.append(try_t)
                c+=1
                
    elif distr is 'uniform':
        a_rands = np.random.random(size=num*dc)*100
        t_rands = np.random.random(size=num*dc)*100
        
    for i in range(num):
        exps.append(lambda  t:a_rands[i]*np.exp(-t/t_rands[i]))
        params.append((a_rands[i], t_rands[i]))
    x = np.linspace(0,dom[0],dom[1])
    y = np.array([f(x) for f in exps]).sum(axis=0)
    return x,y,params

outs = [add_exps(num=1000,distr='uniform',dom=(100,100)) for i in range(10)]
plt.figure(figsize=(12,6))
for i, o in enumerate(outs):
    plt.subplot(1,3,1)
    plt.plot(o[0],o[1], label=i)
    plt.legend()
    plt.subplot(1,3,2)
    plt.hist([a[0] for a in o[2]],histtype='step',alpha=0.6, label=i,lw=3)
    plt.title('alpha')
    plt.legend()
    plt.subplot(1,3,3)
    plt.hist([a[1] for a in o[2]],histtype='step',alpha=0.6, label=i,lw=3)
    plt.title('tau')
    plt.legend()
```
